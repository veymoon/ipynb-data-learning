## Simple Linear Regression

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
mean_x = np.mean(x)
mean_y = np.mean(y)
slope = np.sum((x - mean_x) * (y - mean_y)) / np.sum((x - mean_x) ** 2)
intercept = mean_y - slope * mean_x
print(f"Slope: {slope:.2f}, Intercept: {intercept:.2f}")
```
